EXPLORATORY DATA ANALYSIS - MOSTLY VISULAIZATION

```python
#take the data and create dataframe
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/laptops.csv',encoding = 'latin-1')
df
```

```python
df.info() #It gives the information about the dataframe
```

```python
df.shape #1303 rows ,13 columns
```

```python
df.size #total number of elements in the dataframe
```

```python
#To check if my dataframe has got any null values or not
df.isnull().sum()
```

```python
#Since we have a column named 'Unnamed: 0' which is not of much use, we should remove/drop the 'Unnamed: 0' column
```

```python
df = df.drop(columns = 'Unnamed: 0')
df
```

```python
df.info()
```

```python
#Let's remove 'GB' from the RAM column and keep only the INteger value
df['Ram'] = df['Ram'].str.replace('GB','')
df.info()
```

```python
#COnverting the data of RAM column from string to int
df['Ram'] = df['Ram'].astype('int64')
df.info()
```

```python
#In weight column, remove kg and replace it with empty string
#Convert datatype from string to float
df['Weight'] = df['Weight'].str.replace('kg','')
df['Weight'] = df['Weight'].astype('float64')
df.info()
```

```python
# 1. Take the data and create dataframes
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/CarPrice_Assignment.csv')
df
```

```python
#Visualization
import seaborn as sns 
sns.distplot(df['price'])
# WE came to know that most of the cars are in the price range of 5000 to 18000 euros
```

```python
#We want to consider only numeric datas
#So we can create a dataframe with only numeric data
df_numeric = df.select_dtypes(include = ['float64','int64'])
df_numeric
```

```python
#Now, to remove the column car_ID and symboling, because they are not of much use
df_numeric = df_numeric.drop(['symboling','car_ID'],axis = 1) #axis = 1 -column, axis = 0 - row
df_numeric
```

```python
df_numeric.info()
```

```python
#4.Divide the data into input and output
#Output - price, input - rest of the 13 columns
#df.iloc[row slicing,col slicing]
x = df_numeric.iloc[:,0:13].values
x
```

```python
y = df_numeric.iloc[:,13].values
y
```

```python
#5.Train and Test variables - train_test_split()
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)
#x and y are split into test and train variables
#BY default training variables get 75% of x and y and testing variables are given remaining 25%
```

```python
print(x.shape) #(205, 13) - 100%
print(x_train.shape) #(153, 13) - 75%
print(x_test.shape) #(52, 13) - 25%
```

```python
print(y.shape) #(205, ) - 100%
print(y_train.shape) #(153, ) - 75%
print(y_test.shape) #(52,) -25%
```

```python
# 6.Normalization/Scaling(Done for only input)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
# 7.Run a classifier, regressor or clusterer
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# 8. Fit the model
model.fit(x_train,y_train)
```

```python
# 9.Predict the output
y_pred = model.predict(x_test) #Using input values we predict the output values
y_pred #Predicted output values
```

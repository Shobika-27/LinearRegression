```python
#MACHINE LEARNING - SUPERVISED LEARNING - REGRESSION - LINEAR REGRESSION
#Univariate LinearRegression/Single Linear Regression
#Dataset - Area vs Prices
#Dataset - https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/areavsprices.csv
```

```python
#1.Take the dataset and create dataframe
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/areavsprices.csv')
df
#Imaginary story - A Realestate agent comes to you and gives you the below data and says to create a model for me which could predict the property prices based on the data I gave
# Area is in sqft and prices is in Rupees
```

```python
#2.Preprocessing (filtering) is not required because of limited data
```

```python
#3.Data Visualization - Creation of Graphs
import matplotlib.pyplot as plt
plt.scatter(df['Area'],df['Prices'])
plt.title('Area vs Prices')
plt.xlabel('Area')
plt.ylabel('Prices')
```

```python
#4.Divide the data into input(x) and output(y)
#Input - Area and Output - Prices
#Input(x) is 2-dimentional and Output(y) is always 1-dimentional
#df.iloc[row-slicing,column-slicing]
#In column'splace if there is a ':' then array is 2 dimentional
#.values convert the dataframe into array
x = df.iloc[0:6,0:1].values #df.iloc[:,0:1].values can be also used to retrieve all the values in rows
x
```

```python
type(x)
```

```python
x.ndim
```

```python
y = df.iloc[:,1].values #No ':' in column place so, its 1 dimentional
y
```

```python
df['Prices'].values
```

```python
#5.Train and test step is not required due to limited data
#6.Normalizing the data is also not required
```

```python
#7.Run a classifier, Regressor or clusterer(We apply LInear regressor algorithm)
#sklearn.linear_model - package(collection of libraries), LinearRegression - Library
from sklearn.linear_model import LinearRegression
model = LinearRegression() # whenever we put a library inside a variable then we will be treating it as a method/function
```

```python
#8.FIT the model (Mapping/Plotting the inputs with the output)
model.fit(x,y) #PLotting x,y inside the LINEAR REGRESSION LIBRARY
```

```python
#9.Predict the output
y_pred = model.predict(x) #Using the input values , we predict the outputs
y_pred
```

```python
y #Actual Output values
```

```python
#CONCLUSION : We have to compare the corresponding values of y and y_pred
#So when we compare, we come to know that there is a huge difference
#This huge difference does not mean that the predicted model is wrong
#It means the model is NOT LINEAR/LESS LINEAR 
#LINEARITY of a model depends on nature of data as well as size of the data
```

```python
#INDIVIDUAL PREDICTION
#To know the price for 2000 spft
model.predict([[2000]])
```

```python
#CROSS VERIFY
#y = mx+c (Equation of a straight line)
#m - slope
#c - constant/y-intercept
#y - dependant variable
#x - independant
```

```python
#To find slope(m)
m = model.coef_
m
```

```python
#To find y-intercept/constant(c)
c = model.intercept_
c
```

```python
#y = mx+c
m*2000+c
```

```python
#VISUALIZATION FOR THE BEST FIT LINE
plt.scatter(x,y) #Actual value
plt.plot(x,y_pred,c = 'olive') #Predicted values
plt.title('BEST FIT LINE')
plt.xlabel('Area')
plt.ylabel('Prices')
```
